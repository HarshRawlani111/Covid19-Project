```python
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot =True)
plt.title('Correlation Matrix of Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
#changing data type of string to numeric
df_numerized = df
for col_1 in df.numerized:
    if (df_numerized[col_1].dtype == 'object'):
        df_numerized[col_1] =  df.numerized[col_1].astype('category')
        df_numerized[col_1] =  df.numerized[col_1].cat.codes
df.numerized.head()  
```

```python
#Correlation of all the movie elements
correlation_matrix = df_numerized.corr()
sns.heatmap(correlation_matrix, annot =True)
plt.title('Correlation Matrix of Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
#see by descending order of correlation
correlation_mat = df_numerized.corr()
corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_pair = corr_pairs.sort_values()
sorted_pair 
```

```python
#high Correlation

high_corr = sorted_pair[(sorted_pair)>0.5]
high_corr
```
